### Youtube Rank 인기 콘텐츠 한 페이지 크롤링 => top 100

1. chrome driver 열기
2. 페이지 접속하기 
3. Beautiful Soup 활용하여 html형태로 불러온 데이터를 정리하기
4. 원하는 정보를 찾기
  - 채널 정보를 뭉치 별로 top 100 찾기
5. Excel 파일로 데이터 저장(list -> DataFrame)

````python
from selenium import webdriver
from bs4 import BeautifulSoup

import time    #  time.sleep(대기시간)
import pandas as pd
````

````python
## Colab에선 웹브라우저 창이 뜨지 않으므로 별도 설정한다.
 
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
````

````python
#해당 url로 이동
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=1"
driver.get(url)
````

````python
# 3. BeautifulSoup 활용하여 데이터 정리하기

html = driver.page_source
html
````

````python
soup = BeautifulSoup(html, 'html.parser')
soup
````

4. 원하는 정보 찾기


```
soup.select('태그명')
soup.select('.class속성값')
soup.select('부모태그정보 > 자식태그정보')
soup.select('#id속성값')
```



````python
channel_list = soup.select('tr')
len(channel_list)
````

````python
channel_list = soup.select('.aos-init')
len(channel_list)
````

````python
# 찾은 데이터를 하나씩 살펴보기
channel_list[99]
````

````python
# 채널 100개 -> 1개 선택
channel = channel_list[0]
channel
````

````python
# 채널명 찾기
title = channel.select('h1 > a')
len(title)
````

soup.select() <-> channel.select()

````python
title = channel.select('h1 > a')[0].text.strip() # 공백 제거해줌 \t,\n

title
````

````python
# 카테고리 정보 정리해보기
category = channel.select('p.category')[0].text.strip()
category
````

````python
# 구독자 정보 정리해보기
subscriber = channel.select('td.subscriber_cnt')[0].text.strip()
subscriber
````

````python
# 뷰 수 정보 정리해보기
view = channel.select('td.view_cnt')[0].text.strip()
view
````

````python
# 채널 하나의 정보를 수집
# 1. 채널 선택
channel = channel_list[0] # 블랙핑크

# 2. 1개의 채널 정보를 정리
title = channel.select('h1 > a')[0].text.strip()
category = channel.select('p.category')[0].text.strip()
subscriber = channel.select('td.subscriber_cnt')[0].text.strip()
view = channel.select('td.view_cnt')[0].text.strip()

# 3. 출력하기
print(title, category, subscriber, view)
````

1개의 채널 정보 수집 -> 100개의 정보 수집

````python
### 반복문 활용

for channel in channel_list:

  title = channel.select('h1 > a')[0].text.strip()
  category = channel.select('p.category')[0].text.strip()
  subscriber = channel.select('td.subscriber_cnt')[0].text.strip()
  view = channel.select('td.view_cnt')[0].text.strip()

  print(title, category, subscriber, view)
````

````python
# 리스트로 모든 정보들을 저장하기 \n

results = [] # <- 정보를 저장할 빈 리스트 생성
for channel in channel_list: 

  title = channel.select('h1 > a')[0].text.strip() 
  category = channel.select('p.category')[0].text.strip() 
  subscriber = channel.select('td.subscriber_cnt')[0].text.strip() 
  view = channel.select('td.view_cnt')[0].text.strip() 
  
  data = [title, category, subscriber, view] 
  results.append(data)
  
````

````python
# pandas를 활용하여 엑셀 파일에 저장하기
# 표 모양 DataFrame으로 바꿔주기

df = pd.DataFrame(results)
df
````

````python
df.columns = ['title', 'category', 'subscriber', 'view']
df
````
